Read Dataset

```python
import matplotlib.pyplot as plt
import pandas as pd
```

```python
df_fish = pd.read_csv('fish_data.csv')
```

Split Independent & Dependent

```python
x = df_fish.iloc[:,1:].values
y = df_fish.iloc[:,[0]].values
```

```python
x
```

```python
y
```

```python
y = y.reshape(len(y)) 
```

Split Train Test

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x,y,test_size=0.6,random_state=60)
```

Create Model

```python
from sklearn.ensemble import RandomForestClassifier
```

```python
rf = RandomForestClassifier(n_estimators=100, random_state=36)
```

```python
rf.fit(x_train,y_train)
```

Prediksi

```python
y_prediksi = rf.predict(x_test)
```

Cek Akurasi Dan Evaluasi

```python
from sklearn.metrics import accuracy_score
from sklearn.metrics import classification_report, confusion_matrix
```

```python
print("Nilai Akurasi", accuracy_score(y_test,y_prediksi))
```

```python
print(classification_report(y_test,y_prediksi))
```

```python
cf_matrix = confusion_matrix(y_test, y_prediksi)
cf_matrix
```

Klasifikasi Data Baru

```python
new_data = [[8.38,3.46,0.41]]
new_pred = rf.predict(new_data)
print(new_pred)
```
